```python
import pandas as pd
pd.set_option("max_rows", 100)
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
round(data[data['sex'] == 'Female']['age'].mean())
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
german_amount = data[data['native-country'] == 'Germany']['native-country'].value_counts()
german_amount / data.shape[0] * 100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
filtered_ages = data[data['salary']=='>50K']['age']
print("Mean of age for those who earn more than 50K per year: {0}".format(round(filtered_ages.mean())))
print("Standart deviation of age for those who earn more than 50K per year: +- {0}".format(round(filtered_ages.std(), 1)))
```

```python
other_filtered_data = data[data['salary']=='<=50K']['age']
print("Mean of age for those who earn less than 50K per year: {0}".format(round(other_filtered_data.mean())))
print("Standart deviation of age for those who earn less than 50K per year: +- {0}".format(round(other_filtered_data.std(), 1)))
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
#Short way
pd.crosstab(data['salary'], data['education'])
```

```python
#Long way, but contains answer
high_school_education_type = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters or Doctorate']
people_education_types = data[data['salary']=='>50K']['education'].unique()
result =list(set(high_school_education_type) - set(people_education_types))
print(len(result) == 0)
```

```python
#Третий способ(без использования set)
def is_no_common_data(list1, list2): 
    result = True
    for x in list1: 
        for y in list2:  
            if x == y: 
                result = False
                return result  
                  
    return result 
```

```python
is_no_common_data(people_education_types, high_school_education_type)
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
data.groupby(['race', 'sex'])['age'].describe()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
men = data[data['sex'] == 'Male']
men[men['marital-status'].apply(lambda state: state.startswith('Married'))]['salary'].value_counts()
```

```python
men[men['marital-status'].apply(lambda state: not state.startswith('Married'))]['salary'].value_counts()
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
print("Maximum number of hours a person works per week:")
max_hours_per_week = data['hours-per-week'].max()
print(max_hours_per_week)
```

```python
print('Amount of people who work maximum number per week:')
max_works_people = data[data['hours-per-week'].apply(lambda state: state == max_hours_per_week)]
max_works_people_amount = len(max_works_people)
print(max_works_people_amount)
```

```python
rich_max_works_people = round(max_works_people[max_works_people['salary'] == '>50K'].shape[0] / max_works_people_amount * 100)
print('Percentage of those who earn a lot (>50K) among them: {0}%'.format(rich_max_works_people))
```
